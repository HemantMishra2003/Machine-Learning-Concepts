#### Defining My Change Dataset

```python
import os
import numpy as np
import torch
from torch.utils.data import Dataset
from PIL import Image

class ChangeDetectionDataset(Dataset):

    def __init__(self, A_DIR, B_DIR, L_DIR, transform=None):
        self.A_DIR = A_DIR
        self.B_DIR = B_DIR
        self.L_DIR = L_DIR
        self.files = sorted([f for f in os.listdir(A_DIR) if f.endswith(('.png','.jpg','.jpeg'))])
        self.transform = transform

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):

        f = self.files[idx]

        imgA = np.array(Image.open(os.path.join(self.A_DIR, f)).convert("RGB"))
        imgB = np.array(Image.open(os.path.join(self.B_DIR, f)).convert("RGB"))
        mask = np.array(Image.open(os.path.join(self.L_DIR, f)))

        mask = (mask > 128).astype(np.uint8)

        image = np.concatenate([imgA, imgB], axis=2)

        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented["image"]
            mask = augmented["mask"]

        image = torch.tensor(image).permute(2,0,1).float()
        mask = torch.tensor(mask).float()

        return image, mask
```

```python
encoder_list = ['resnet34', 'efficientnet-b0', 'mobilenet_v2']
# here we are choosing three different pretrained model
# which is helpful for feature extraction
models = []

for encoder_name in encoder_list:
    model = smp.Unet(  # here we are using unet model best for segmentation
        encoder_name=encoder_name,
        encoder_weights='imagenet',
        in_channels=6, # 6 channel ,since have two iamage
        classes=1,
        activation='sigmoid'
    ).to(device)

    models.append(model)
```

#### Training Pipeline

##### Defining Training Setup

```python
train_dataset = ChangeDetectionDataset(
    A_DIR="/content/EarthData/GoogleEartth/Train/T1",
    B_DIR="/content/EarthData/GoogleEartth/Train/T2",
    L_DIR="/content/EarthData/GoogleEartth/Train/Label"
)

from torch.utils.data import DataLoader

train_loader = DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True
)
```

```python
from tqdm import tqdm

criterion = nn.BCELoss()

optimizers = []
for m in models:
    optimizers.append(torch.optim.Adam(m.parameters(), lr=1e-4))

for epoch in range(15):
    print(f"Starting Epoch: {epoch+1}")

    total_loss = 0
    num_batches = 0

    for images, masks in tqdm(train_loader):
        images = images.to(device)
        masks = masks.to(device).unsqueeze(1).float()

        # Sirf ek model per batch (cycle through models)
        model_idx = num_batches % len(models)  # 0,1,2,0,1,2,...
        model = models[model_idx]
        optimizer = optimizers[model_idx]

        optimizer.zero_grad()
        preds = model(images)
        loss = criterion(preds, masks)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        num_batches += 1

    avg_loss = total_loss / num_batches
    print(f"Epoch {epoch+1} | Loss: {avg_loss:.4f}")
```

```python
import numpy as np
from PIL import Image
import torch
from torchvision.transforms import functional as F  # yeh resize ke liye

def predict_change_any_size(before_path, after_path, models, device, strong_thresh=0.25):
    """
    Kisi bhi size ki images → agar mismatch hai toh automatically resize kar deta hai
    """
    # Load images
    before_pil = Image.open(before_path).convert('RGB')
    after_pil  = Image.open(after_path).convert('RGB')

    before_size = before_pil.size  # (W, H)
    after_size  = after_pil.size

    if before_size != after_size:
        print(f"⚠️ Size mismatch → Before: {before_size[::-1]}, After: {after_size[::-1]}")

        # Bigger resolution wali image ka size target banao
        target_size = max(before_size, after_size, key=lambda s: s[0]*s[1])  # area ke hisaab se bigger

        print(f"→ Resizing smaller image to {target_size[::-1]}")

        if before_size != target_size:
            before_pil = F.resize(before_pil, target_size[::-1])  # (H, W) format mein deta hai resize
        if after_size != target_size:
            after_pil = F.resize(after_pil, target_size[::-1])

    # Ab numpy array mein convert
    before_img = np.array(before_pil)
    after_img  = np.array(after_pil)

    H, W = before_img.shape[:2]

    # 6-channel input
    test_image = np.concatenate([before_img, after_img], axis=2).astype(np.float32)

    # Normalize (ImageNet mean & std — agar model inpe train hua hai)
   # Yeh line replace kar do purani normalization wali line se
    mean = np.array([0.485, 0.456, 0.406] * 2)   # → shape (6,)
    std  = np.array([0.229, 0.224, 0.225] * 2)   # → shape (6,)

    test_image = (test_image / 255.0 - mean) / std
    # Torch tensor: (1, 6, H, W)
    test_tensor = torch.from_numpy(test_image.transpose(2, 0, 1)).unsqueeze(0).float().to(device)

    # Ensemble
    p_final = torch.zeros((1, 1, H, W), device=device)
    with torch.no_grad():
        for model in models:
            pred = model(test_tensor)
            p_final += pred

    p_final /= len(models)

    # Mask
    change_mask = (p_final[0, 0] >= strong_thresh).cpu().numpy().astype(np.uint8)

    # Overlay
    overlay = after_img.copy()
    overlay[change_mask == 1] = [255, 0, 0]

    bw_mask_pil = Image.fromarray((change_mask * 255).astype(np.uint8))
    overlay_pil = Image.fromarray(overlay)

    return bw_mask_pil, overlay_pil, change_mask


# ---------------------- Use like this ----------------------

before_path = "/content/T1AA.jpeg"
after_path  = "/content/T2BB.jpeg"

bw_mask, red_overlay, raw_mask = predict_change_any_size(
    before_path,
    after_path,
    models=models,
    device=device,
    strong_thresh=0.25   # ya 0.3 / 0.35 try kar sakta hai
)

bw_mask.save("DL_ENSEMBLE_change_map_fixed.png")
red_overlay.save("DL_ENSEMBLE_red_highlight_fixed.png")

print("✅ Ab size mismatch ka tension khatam!")
print(f"Change pixels: {raw_mask.sum()} / {raw_mask.size} ≈ {raw_mask.mean()*100:.2f}% area changed")
```

```python
import os
import numpy as np
from PIL import Image
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, jaccard_score
from tqdm import tqdm  # progress bar ke liye (nice dikhega)

# ------------------ Settings ------------------
test_dir = "/content/test/"          # tera main test folder
before_dir = os.path.join(test_dir, "T1")     # ya "before"
after_dir  = os.path.join(test_dir, "T2")     # ya "after"
gt_dir     = os.path.join(test_dir, "label")  # ya "gt" ya "mask"

strong_thresh = 0.25   # tera threshold

# Lists to store metrics for average
all_metrics = {"acc": [], "prec": [], "rec": [], "f1": [], "iou": []}

# Get all image names from before folder (assume same in all folders)
image_names = sorted([f for f in os.listdir(before_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])

print(f"Found {len(image_names)} images in test set. Evaluating...\n")

for img_name in tqdm(image_names):
    before_path = os.path.join(before_dir, img_name)
    after_path  = os.path.join(after_dir, img_name)
    gt_path     = os.path.join(gt_dir, img_name)          # same name

    # Agar gt mein extension alag hai (jaise .png instead of .jpg) toh yeh try kar
    # gt_path = os.path.join(gt_dir, os.path.splitext(img_name)[0] + ".png")

    if not os.path.exists(gt_path):
        print(f"Skipping {img_name} — ground truth nahi mila!")
        continue

    # Tera prediction call
    bw_mask_pil, red_overlay_pil, pred_mask = predict_change_any_size(
        before_path,
        after_path,
        models=models,
        device=device,
        strong_thresh=strong_thresh
    )

    # Ground truth load + binarize
    gt_pil = Image.open(gt_path).convert('L')  # grayscale
    gt_mask = np.array(gt_pil)
    gt_mask = (gt_mask > 127).astype(np.uint8)  # 255 → 1, 0 → 0

    # Sizes match kar lo (agar nahi to resize — lekin usually same hone chahiye)
    if pred_mask.shape != gt_mask.shape:
        print(f"Size mismatch in {img_name}! Skipping metrics.")
        continue

    # Flatten for sklearn
    pred_flat = pred_mask.flatten()
    gt_flat   = gt_mask.flatten()

    # Metrics calculate
    try:
        acc  = accuracy_score(gt_flat, pred_flat)
        prec = precision_score(gt_flat, pred_flat, zero_division=0)
        rec  = recall_score(gt_flat, pred_flat, zero_division=0)
        f1   = f1_score(gt_flat, pred_flat, zero_division=0)
        iou  = jaccard_score(gt_flat, pred_flat, zero_division=0)

        all_metrics["acc"].append(acc)
        all_metrics["prec"].append(prec)
        all_metrics["rec"].append(rec)
        all_metrics["f1"].append(f1)
        all_metrics["iou"].append(iou)

        print(f"\nImage: {img_name}")
        print(f"Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f} | IoU: {iou:.4f}")

    except Exception as e:
        print(f"Error in {img_name}: {e}")

# Final average
if all_metrics["acc"]:
    print("\n" + "="*50)
    print("OVERALL TEST SET RESULTS (Average over all images)")
    print(f"Accuracy  : {np.mean(all_metrics['acc']):.4f}")
    print(f"Precision : {np.mean(all_metrics['prec']):.4f}")
    print(f"Recall    : {np.mean(all_metrics['rec']):.4f}")
    print(f"F1-Score  : {np.mean(all_metrics['f1']):.4f}")
    print(f"IoU       : {np.mean(all_metrics['iou']):.4f}")
    print("="*50)
else:
    print("Koi bhi image evaluate nahi hui — paths check kar bhai!")
```
